```python
import pandas as pd
import numpy as np
data=[
   ["rajan",26,40000],
   ["sanju",23,50000],
   ["jyothi",34,60000],
   ["santhi",np.nan,25000],
   ["sravs",25,700000],
   ["chinna",np.nan,30000]
]
df1=pd.DataFrame(data,columns=["Name","Age","Salary"])
print(df1)
m=df1['Age'].mean()
df1['Age']=df1['Age'].fillna(m)
print()
print(df1)
```

```python
import pandas as pd
import numpy as np
data=[
   ["rajan",26,40000],
   ["sanju",23,50000],
   ["jyothi",34,60000],
   ["santhi",np.nan,25000],
   ["sravs",25,700000],
   ["chinna",np.nan,30000]
]
df1=pd.DataFrame(data,columns=["Name","Age","salary"])
print(df1)
df2=df1.replace(np.nan,0)
print()
print(df2)
```
