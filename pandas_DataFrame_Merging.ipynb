```python
import pandas as pd
d1={
    "id":[1,2,3,4,5,6],
    "Name":["pradhan","venu","madurima","nireekshan","sanju","veeru"],
    "subject":["English","java","html","python","c","dot net"]
}
d2={
    "id":[11,12,13,14,15,16],
    "Name":["srinu","sumanth","neelima","daniel","Arjun","veeru"],
    "subject":["java","html","cpp","python","c","dot net"]
}

df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
inn_join=pd.merge(df1,df2,on="subject",how="inner")
print(df1)
print()
print(df2)
print()
print(inn_join)
```

```python
import pandas as pd
d1={
    "employee":["nandu","nani","sathya","lavanya"],
    "Group":["development","HR","testing","testing"]
}
d2={
    "employee":["nani","nandu","sathya","lavanya"],
    "Hire_date":[2010,2012,2014,2016]
}
d3={
    "Group":["testing","development","HR"],
    "supervisor":["shafi","sanju","santhi"]
}
df1=pd.DataFrame(d1)
df2=pd.DataFrame(d2)
df3=pd.DataFrame(d3)

one_one=pd.merge(df1,df2)
many_one=pd.merge(one_one,df3)

print(df1)
print()
print(df2)
print()
print(many_one)
```
